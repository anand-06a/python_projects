```python
# <----------------->
#<---vhicle service management sydtem--->
#<------------------>

service_menu={
    1:("oil change",1200),
    2:("engine check",2500),
    3:("brake service",1800),
    4:("wheel alignment",1000),
    5:("car wash",500)
}

class vehicleservice:
    def register_customer(self):
        print("\n=== customer details ===")
        name=input("enter customer name:")
        phone=(input("enter phone number:"))
        vehicle=input("enter vehicle name:")
        number=(input("enter vehicle number:"))
        
        return name,phone,vehicle,number
    def select_service(self):
        total=0
        selected_services=[]
        while True:
            print("\n----service menu----")
            for key,value in service_menu.items():
                print(f"{key}. {value[0]} -₹{value[1]}")
            print("0.finish")
            choice=int(input("select service:"))
            if choice==0:
                break
            elif choice in service_menu:
                service,price=service_menu[choice]
                selected_services.append(service)
                total+=price
                print(service,"added")
            else:
                print("invalid choice")
        return selected_services,total
    def generate_bill(self,customer):
        name,phone,vehicle,number=customer
        services,total=self.select_service()
        print("\n====bill====")
        print("customer:",name)
        print("phone:",phone)
        print("vehicle:",vehicle)
        print("number:",number)
        print("\n services:")
        for item in services:
            print("-",item)
        
        print("\n total amount:₹",total)
        file=open("service_records.txt","a")
        
        file.write("\n----------------\n")
        file.write("customer : " + name + "\n")
        file.write("phone : " + phone + "\n")
        file.write("vehicle : " + vehicle + "\n")
        file.write("number : " + number + "\n")
        file.write("services : " + ", ".join(services) + "\n")
        file.write("total : " + str(total) + "\n")
        
        file.close()
        print("\n record saved successfully.")
    def view_records(self):
        print("\n ==== service records ==== \n")
        try:
            with open("service_recotds.txt","r",encoding="utf-8") as file:
                print(file.read())
                file.close()
        except filenotfounderror:
            print("no records found.")
obj = vehicleservice()
while True:
    print("\n ==============================")
    print(" vehicle service management ")
    print("==============================")
    print("1. new service")
    print("2. view records")
    print("3. exit")
    choice = input("enter choice: ")
    if choice == "1":
        customer = obj.register_customer()
        obj.generate_bill(customer)
    elif choice == "2":
        obj.view_records()
    elif choice == "3":
        print("thank you")
        break
    else:
        print("invalid choice")            

            
```
